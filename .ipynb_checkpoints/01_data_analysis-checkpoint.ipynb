## 2.2 Target

Begin by exploring the target distribution

```python
# Count the occurrences of each value in the 'exited' column
exited_counts = df['Exited'].value_counts()

# Extract labels and colors based on the unique values
labels = exited_counts.index.tolist()
colors = ['#66b3ff', '#ff9999'][:len(labels)]

# Calculate percentages
total = exited_counts.sum()
percentages = [(count / total) * 100 for count in exited_counts]

# Create a bar plot
plt.figure(figsize=(8, 6))
bars = plt.bar(labels, exited_counts, color=colors)

# Add labels on top of the bars
for bar, count, percent, label in zip(bars, exited_counts, percentages, labels):
    label_text = f'{label}: ({percent:.1f}%)'
    plt.text(bar.get_x() + bar.get_width() / 2, count, label_text,
             ha='center', va='bottom', fontsize=12)

# Add title and axis labels
plt.title('Distribution of Exited', fontsize=16)
plt.xlabel('Outcome', fontsize=14)
plt.ylabel('Count', fontsize=14)

# Show the plot
plt.show()
```

- We found that the exited rate was 79.6% and the not exited rate was 20.4%. This **unbalanced distribution** should be taken into account when evaluating classification model performance.
We can see this is **imbalanced data**.
- Handle Imbalanced Data For a Classification Problem:
  a. Choose Proper Evaluation Metric
  b. Resampling (Oversampling and Undersampling)
  c. Searching Optimal Value From a Grid
  d. Modeling Techniques: Some machine learning algorithms can deal better with imbalanced data. (Tree based algorithm)

- We should **test**

```python
def plot_frequency(data, column, title, xlabel):
    plt.figure(figsize=(10, 6))
    unique_values = data[column].value_counts()
    
    sns.barplot(x=unique_values.index, y=unique_values.values, hue=unique_values.index, 
            palette='viridis', legend=False)
    
    plt.title(title, fontsize=16)
    plt.xlabel(xlabel, fontsize=14)
    plt.ylabel('Frequency', fontsize=14)
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.show()

# Plot frequency for each column
plot_frequency(df, 'Geography', 'Frequency of Geography', 'Geography')
plot_frequency(df, 'Gender', 'Frequency of Gender', 'Gender')
```

## 2.5.b Rare labels:

Investigate now if there are labels that are present only in a small number of bet and parameter:

```python
def analyse_rare_labels(df, var, rare_perc):
    df = df.copy()

    # determine the % of observations per category
    tmp = df.groupby(var)['Exited'].count() / len(df)

    # return categories that are rare
    return tmp[tmp < rare_perc]

# print categories that are present in less than 1 % of the observations

for var in cat_vars:
    print(analyse_rare_labels(df, var, 0.01))
    print()
```

```python
df.info()
```

```python
# Selecting numerical columns
numeric_columns = df.drop(columns=['RowNumber', 'CustomerId', 'Surname', 'Geography', 'Gender',]).select_dtypes(include=[np.number])

# Calculating the correlation matrix
correlation = numeric_columns.corr()

# Creating a heatmap to understand the correlation
plt.figure(figsize=(16, 9))
sns.heatmap(correlation, cbar=True, square=True,
            fmt='.1f', annot=True, annot_kws={'size': 8},
            cmap='Oranges')

# Showing the plot
plt.show()
```
